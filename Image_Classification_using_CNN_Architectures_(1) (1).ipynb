## (1) What is a Convolutional Neural Network (CNN), and how does it differ from traditional fully connected neural networks in terms of architecture and performance on image data?

Answer:
A Convolutional Neural Network (CNN) is a specialized type of deep learning model primarily designed for processing and analyzing visual data, such as images or videos. CNNs automatically learn spatial hierarchies of features—from edges and textures to complex shapes—directly from input images.

### 1. Architecture of CNN:
CNNs are made up of several key layers:

### (1) Convolutional Layer:
- erforms convolution operations using filters (kernels) that slide over the image to extract local features (like edges, corners, or color gradients).

- Each filter learns a specific feature automatically.

- Output of this layer is called a feature map.

### (2) Activation Function (usually ReLU):
- Adds non-linearity to help the network learn complex patterns.

### (3) Pooling Layer:
- Reduces the spatial size of the feature maps (downsampling) while keeping important features intact.

- Common types: Max Pooling and Average Pooling.

- Helps make the model more computationally efficient and reduces overfitting.

### (4) Fully Connected (Dense) Layer:
- After several convolution + pooling layers, the output is flattened and fed into fully connected layers for classification (e.g., predicting image labels).

## (2) Discuss the architecture of LeNet-5 and explain how it laid the foundation for modern deep learning models in computer vision. Include references to its original research paper.

Answer:
- (1) LeNet-5 is one of the earliest and most influential Convolutional Neural Network (CNN) architectures, developed by Yann LeCun and his colleagues in 1998. It was introduced in their famous research paper titled “Gradient-Based Learning Applied to Document Recognition” (LeCun et al., 1998). This model was primarily designed for handwritten digit recognition tasks, such as reading ZIP codes and bank check numbers. LeNet-5 demonstrated the capability of neural networks to automatically learn visual features directly from raw image pixels, which became the foundation of modern computer vision models.

- (2)The architecture of LeNet-5 consists of seven layers with trainable parameters, excluding the input layer. The network accepts a grayscale image of size 32×32 pixels as input. The first layer, known as C1, is a convolutional layer that applies six filters of size 5×5 to the input image. This layer detects simple visual features such as edges and corners. The next layer, S2, is a subsampling or pooling layer that performs average pooling with a 2×2 filter, reducing the spatial size of the feature maps and making the representation more compact and invariant to small translations.

- (3)After pooling, the third layer, C3, is another convolutional layer with sixteen feature maps. This layer combines information from different subsets of the previous layer and learns more complex features. It is followed by another pooling layer, S4, which again performs average pooling to further reduce the spatial dimensions of the data. The fifth layer, C5, is a convolutional layer with 120 feature maps using 5×5 kernels. Since the feature maps at this stage are very small, this layer is effectively fully connected to the previous one. It acts as a bridge between the convolutional feature extraction layers and the final classification layers.

- (4)Next, the F6 layer is a fully connected layer with 84 neurons, which uses the sigmoid or tanh activation function. This layer helps in learning abstract combinations of the extracted features. Finally, the output layer consists of ten neurons corresponding to the ten possible classes of handwritten digits (0–9). The softmax function is used here to produce probability scores for each class, allowing the network to classify the input image into one of the digit categories.

- (5)LeNet-5 introduced several key concepts that became the building blocks of modern CNNs. These include local receptive fields, which allow neurons to process only small regions of the input image; shared weights, which use the same filters across the entire image to reduce the number of parameters; and subsampling (pooling), which helps achieve translation invariance and reduces overfitting. Additionally, LeNet-5 utilized gradient-based learning (backpropagation) for training, allowing the network to automatically learn features from data rather than relying on manual feature engineering.

- (6)This architecture laid the foundation for modern deep learning models in computer vision. The core design principles of LeNet-5—convolutional layers for feature extraction, pooling layers for dimensionality reduction, and fully connected layers for classification—are still present in contemporary networks such as AlexNet, VGGNet, ResNet, and Inception. Modern architectures build upon LeNet-5 by adding more layers, using ReLU activation functions, introducing batch normalization, dropout, and leveraging GPU acceleration for large-scale data training. Nevertheless, the fundamental ideas proposed by LeNet-5 remain central to all CNN-based computer vision systems today.

- (7) summary, LeNet-5 was the pioneering CNN that demonstrated how deep networks could automatically learn hierarchical representations from image data. It marked the beginning of deep learning applications in image recognition and inspired the development of all major convolutional architectures that followed.

## (3) Compare and contrast AlexNet and VGGNet in terms of design principles,number of parameters, and performance. Highlight key innovations and limitations of each.

Answer:
### (1) Design Principles:
- AlexNet (2012): Introduced deep CNNs with 8 layers (5 convolutional + 3 fully connected). Used large filters (11×11, 5×5) and ReLU activation.

- VGGNet (2014): Used very deep networks (16–19 layers) with small 3×3 filters for simplicity and better feature extraction.

### (2) Number of Parameters:
- AlexNet: ~60 million parameters.

- VGGNet: ~138 million parameters (much larger, more memory required).

### (3) Performance:
- AlexNet: Achieved 15.3% top-5 error on ImageNet; first to use GPUs for training.

- VGGNet: Achieved higher accuracy than AlexNet but slower due to its size and computation cost.

### (4) Key Innovations:
- AlexNet: Introduced ReLU activation, dropout, and GPU training.

- VGGNet: Introduced stacked small filters (3×3) and deeper architecture for better accuracy.

### (5) Limitations:
- AlexNet: Shallow network, large filters, prone to overfitting.

- VGGNet: Too large and computationally heavy; difficult to train on limited hardware.


## (4) What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data.

Answer:
### (1) Definition:
Transfer learning is a technique where a pre-trained model (trained on a large dataset like ImageNet) is reused for a new but related task. Instead of training from scratch, the model’s learned features are transferred to the new dataset.

### (2) How It Works:
- The lower layers of a CNN (which learn general features like edges or textures) are kept or “frozen.”

- The upper layers are retrained or fine-tuned for the specific image classification task.

### (3) Reduces Computational Cost:
- Saves training time since most of the network is already trained.

- Requires less hardware and data because only a few layers are retrained instead of the entire network.

### (4) Improves Model Performance with Limited Data:
- Pre-trained models already contain rich, general features learned from large datasets.

- This helps achieve high accuracy even when the new dataset is small or lacks diversity.

Example:

Using a pre-trained VGG16 or ResNet model on ImageNet and fine-tuning it for a custom dataset like flower or animal classification.

## (5) Describe the role of residual connections in ResNet architecture. How do they address the vanishing gradient problem in deep CNNs?

Answer:
### (1) ResNet Introduction:
-  (Residual Network) was introduced by He et al. (2015) to allow training of very deep CNNs (up to hundreds of layers).

- It introduced the concept of residual connections (skip connections).

### (2) Role of Residual Connections:
- Residual connections skip one or more layers and add the input of a layer directly to its output:

- Output=F(x)+x

- Here, 𝐹(𝑥) F(x) is the learned function, and 𝑥 x is the original input.

- This helps the network learn only the residual mapping instead of the full transformation.

### (3) Addressing the Vanishing Gradient Problem:
- In very deep networks, gradients can become extremely small (vanish) during backpropagation, preventing effective learning.

- Residual connections allow gradients to flow directly through skip paths, ensuring stronger gradient signals reach earlier layers.

- This stabilizes training and enables much deeper models to converge.

### (4) Key Benefits:
- Prevents vanishing/exploding gradients.

- Enables faster convergence and better accuracy.

- Allows effective training of networks with 50, 101, or even 152 layers.

```python
import time
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# 1) Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2) Preprocess: add channel, cast to float, resize to 32x32, normalize
x_train = np.expand_dims(x_train, -1).astype('float32')  # (60000, 28, 28, 1)
x_test  = np.expand_dims(x_test,  -1).astype('float32')  # (10000, 28, 28, 1)

# Resize each image to 32x32 (tf.image.resize handles batches)
x_train = tf.image.resize(x_train, [32, 32]).numpy() / 255.0
x_test  = tf.image.resize(x_test,  [32, 32]).numpy() / 255.0

# 3) Build LeNet-5-like model
model = models.Sequential([
    layers.Conv2D(6, kernel_size=(5,5), activation='tanh', input_shape=(32,32,1)),   # C1
    layers.AveragePooling2D(pool_size=(2,2), strides=(2,2)),                         # S2 (pool_size specified)
    layers.Conv2D(16, kernel_size=(5,5), activation='tanh'),                        # C3
    layers.AveragePooling2D(pool_size=(2,2), strides=(2,2)),                         # S4 (pool_size specified)
    layers.Flatten(),
    layers.Dense(120, activation='tanh'),                                            # C5 ~ FC
    layers.Dense(84, activation='tanh'),                                             # F6
    layers.Dense(10, activation='softmax')                                           # Output
])

# 4) Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 5) Train and time it
start_time = time.time()
history = model.fit(x_train, y_train,
                    epochs=5,
                    batch_size=128,
                    validation_data=(x_test, y_test),
                    verbose=1)
end_time = time.time()

# 6) Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\n Test Accuracy: {test_acc*100:.2f}%")
print(f" Training Time: {end_time - start_time:.2f} seconds")
```

```python
import tensorflow as tf
import tensorflow_datasets as tfds

# Load the tf_flowers dataset with labels
dataset, info = tfds.load('tf_flowers', with_info=True, as_supervised=True)
train_ds = dataset['train']

# Check dataset info
print("Classes:", info.features['label'].names)
print("Total Images:", info.splits['train'].num_examples)
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

def preprocess(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = image / 255.0
    return image, label

# Split dataset (80% training, 20% validation)
train_size = int(0.8 * info.splits['train'].num_examples)

train_ds = train_ds.take(train_size)
val_ds = dataset['train'].skip(train_size)

# Apply preprocessing
train_ds = train_ds.map(preprocess).shuffle(1000).batch(BATCH_SIZE)
val_ds = val_ds.map(preprocess).batch(BATCH_SIZE)
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models

# Load VGG16 without top classifier layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))

# Freeze base layers initially
for layer in base_model.layers:
    layer.trainable = False

# Add custom classifier
model = models.Sequential([
    base_model,
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(info.features['label'].names), activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
import time
start_time = time.time()

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)

end_time = time.time()

print(f"\n Training completed in {end_time - start_time:.2f} seconds")
```

```python
val_loss, val_acc = model.evaluate(val_ds)
print(f" Validation Accuracy: {val_acc*100:.2f}%")
```

```python
# -----------------------------
# Import required libraries
# -----------------------------
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Step 1: Build a simplified AlexNet model
# -----------------------------
def build_alexnet():
    model = models.Sequential([
        layers.Conv2D(96, (11, 11), strides=4, activation='relu', input_shape=(227, 227, 3)),
        layers.MaxPooling2D((3, 3), strides=2),
        layers.Conv2D(256, (5, 5), padding='same', activation='relu'),
        layers.MaxPooling2D((3, 3), strides=2),
        layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
        layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
        layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D((3, 3), strides=2),
        layers.Flatten(),
        layers.Dense(4096, activation='relu'),
        layers.Dense(4096, activation='relu'),
        layers.Dense(1000, activation='softmax')
    ])
    return model

model = build_alexnet()
model.summary()

# -----------------------------
# Step 2: Upload an example image
# -----------------------------
print("📁 Upload an image (e.g. cat, dog, etc.) to visualize feature maps:")
uploaded = files.upload()  # upload any image

for fn in uploaded.keys():
    img_path = fn

# -----------------------------
# Step 3: Preprocess the image
# -----------------------------
img = image.load_img(img_path, target_size=(227, 227))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

plt.imshow(img)
plt.title("Input Image")
plt.axis("off")
plt.show()

# -----------------------------
# Step 4: Visualize filters of the first Conv layer
# -----------------------------
first_conv = model.layers[0]
filters, biases = first_conv.get_weights()
print("Filter shape:", filters.shape)

# Normalize filter values for display
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)

# Plot first 6 filters
n_filters = 6
plt.figure(figsize=(15, 5))
for i in range(n_filters):
    f = filters[:, :, :, i]
    plt.subplot(1, n_filters, i+1)
    plt.imshow(f[:, :, 0], cmap='viridis')
    plt.axis('off')
plt.suptitle("First Conv Layer Filters (AlexNet)")
plt.show()

# -----------------------------
# Step 5: Visualize feature maps (activations)
# -----------------------------
# Create a new model that outputs feature maps from the first conv layer
feature_extractor = tf.keras.Model(inputs=model.inputs, outputs=model.layers[0].output)
feature_maps = feature_extractor.predict(img_array)

print("Feature map shape:", feature_maps.shape)

# Plot first 6 feature maps
plt.figure(figsize=(15, 5))
for i in range(6):
    plt.subplot(1, 6, i+1)
    plt.imshow(feature_maps[0, :, :, i], cmap='viridis')
    plt.axis('off')
plt.suptitle("Feature Maps from First Conv Layer")
plt.show()
```

```python
# Import required libraries
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np


# Step 1: Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize pixel values (0–255 -> 0–1)
x_train, x_test = x_train / 255.0, x_test / 255.0


# Step 2: Resize images to fit InceptionV3 (75x75x3)
x_train_resized = tf.image.resize(x_train, (75, 75))
x_test_resized = tf.image.resize(x_test, (75, 75))


# Step 3: Load pre-trained InceptionV3 (GoogLeNet variant)
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(75, 75, 3))

# Freeze base layers to retain pre-trained features
for layer in base_model.layers:
    layer.trainable = False


# Step 4: Add custom classification layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
x = Dense(256, activation='relu')(x)
predictions = Dense(10, activation='softmax')(x)  # CIFAR-10 has 10 classes

model = Model(inputs=base_model.input, outputs=predictions)


# Step 5: Compile the model
model.compile(optimizer=Adam(learning_rate=0.0005),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


# Step 6: Train the model
history = model.fit(
    x_train_resized, y_train,
    validation_data=(x_test_resized, y_test),
    epochs=10,
    batch_size=64,
    verbose=1
)


# Step 7: Evaluate the model
test_loss, test_acc = model.evaluate(x_test_resized, y_test, verbose=0)
print(f"✅ Test Accuracy: {test_acc * 100:.2f}%")


# Step 8: Plot training vs validation accuracy
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training vs Validation Accuracy (InceptionV3 on CIFAR-10)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
